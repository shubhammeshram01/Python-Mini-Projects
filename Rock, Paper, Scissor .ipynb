# Rock, Paper, Scissor :

```python
import random
```

```python
print("**********Welcome To The Game**********")
print("Choose any number from below :\n1. Rock\n2. Paper\n3. Scissor")

user_score = 0
comp_score = 0
tie_score = 0
while True:
    u_choice = int(input("Choose any number : "))
    while u_choice<1 or u_choice>3:
        u_choice = int(input("Choose valid number : "))
        print(f"The user choice is = {u_choice}")
    if u_choice==1:
        user_choice="Rock"
    elif u_choice==2:
        user_choice="Paper"
    else:
        user_choice="Scissor"
    print("The user choice is = {}".format(user_choice))
    print()
    print("**********Now it's computer turn**********")
    c_choice = random.randint(1,3)
    print(f"The computer choice is = {c_choice}")
    if c_choice==1:
        comp_choice="Rock"
    elif c_choice==2:
        comp_choice="Paper"
    else:
        comp_choice="Scissor"
    print(f"The computer choice is = {comp_choice}")
    
    if ((user_choice=="Rock" and comp_choice=="Paper") or (user_choice=="Paper" and comp_choice=="Rock")):
        print("Paper Wins")
        result = "Paper"
    elif ((user_choice=="Rock" and comp_choice=="Scissor") or (user_choice=="Scissor" and comp_choice=="Rock")):
        print("Rock Wins")
        result = "Rock"
    elif ((user_choice=="Scissor" and comp_choice=="Paper") or (user_choice=="Paper" and comp_choice=="Scissor")):
        print("Scissor Wins")
        result = "Scissor"
    
    if result == user_choice:
        user_score += 1
    elif result == comp_choice:
        comp_score += 1
    else:
        tie_score += 1
    print()
    print(f"The user score is = {user_score}")
    print(f"The computer score is = {comp_score}")
    print(f"The tie score is = {tie_score}")
    
    repeat = input("Do you wanna play again ? (Yes/No)")
    if repeat == "No" or repeat == "no" :
        break
    print("*"*100)
print("Game is Over....")
print("Thanks for Playing....")
```
